```python
#importing libraries for reading csv file
import pandas as pd
# math for floor functions used later
import math

df=pd.read_csv('titanic/train.csv')
df.head()
```

```python
#dropping unnecessary columns
df.drop(['PassengerId','Name','Cabin','Ticket','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
#declaring dependent(output) and independent(parameters) variable fields
output=df.Survived
parameters=df.drop('Survived',axis='columns')
```

```python
#Using dummy to make text parameters to number parameters
dummy= pd.get_dummies(parameters.Sex)
parameters=pd.concat([parameters,dummy],axis='columns')
#dropping the column after converting
parameters=parameters.drop('Sex',axis='columns')
parameters.head()
```

```python
#creating a newcolumn to add number of siblings, parents and children in fam size
df2= pd.DataFrame(parameters)
df2['Family_Size']=df2['SibSp']+df2['Parch']
parameters.head()
```

```python
#dropping the two colums after adding
parameters=parameters.drop(['SibSp','Parch'],axis='columns')
parameters.head()
```

```python
#checking null values
parameters.columns[parameters.isna().any()]
```

```python
#found that age has na values.seeing to change it.
parameters.head(10)
```

```python
#filling na values with mean values and using floor function to get integer values
parameters.Age = parameters.Age.fillna(math.floor(parameters.Age.mean()))
parameters.head(10)
```

```python
#using sklearn to split datasets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(parameters,output,test_size=0.2)
```

```python
#length of training data set(80%)
len(x_train)
```

```python
#length of testing dataset(20%)
len(x_test)
```

```python
#total length
len(parameters)
```

```python
#using gaussian algorithm
from sklearn.naive_bayes import GaussianNB
algorithm=GaussianNB()
```

```python
#fitting the algorithm
algorithm.fit(x_train,y_train)
```

```python
#accuracy of the algorithm
algorithm.score(x_test,y_test)
```

```python
#true values of testing data
y_test.head()
```

```python
#predicted values of testing data
algorithm.predict(x_test)
```

```python
#probability of not surviving and surviving for each person in test data
algorithm.predict_proba(x_test)
```
